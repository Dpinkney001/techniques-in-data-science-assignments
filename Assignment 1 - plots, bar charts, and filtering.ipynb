```python
#Duvall Pinkney  
#Duvall.Pinkney@gmail.com
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
storeData = pd.read_csv("DOHMH_New_York_City_Restaurant_Inspection_Results.csv")
```

```python
storeData
```

```python
storeData["INSPECTION DATE"] = pd.to_datetime(storeData["INSPECTION DATE"])
```

```python
storeData.head()
```

```python
storeData["SCORE"].hist(bins = 30)
plt.title("Distribution of restaurant inspection scores")
plt.xlabel("Score")
plt.ylabel("# of resturants")
```

```python
borough_counts = storeData["BORO"].value_counts()
borough_counts.plot.bar()
plt.title("Borough restaurant inspections")
plt.xlabel("Borough")
plt.ylabel("# of inspections")
```

```python
# mean of the january scores
jan_filter = storeData["INSPECTION DATE"].dt.month == 1
storeData[jan_filter]["SCORE"].mean()
```

```python
# mean of the july scores
july_filter = storeData["INSPECTION DATE"].dt.month == 7
storeData[july_filter]["SCORE"].mean()
```

```python
food_counts = storeData["CUISINE DESCRIPTION"].value_counts()
food_counts
```

```python
food_counts.head(15).plot.bar()
plt.title("Top 15 inspected cuisines")
plt.xlabel("Cuisine")
plt.ylabel("# of inspections")
```
